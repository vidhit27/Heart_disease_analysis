```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df.head()
```

cp->chestpain(0-typical angina, 1-atypical angina, 2-non-anginal pain, 3-asymptomatic )    
trestbps->resting blood pressure (mm Hg)     
chol->serum cholestrol (mg/dl)   
fbs->fasting blood suger (>120 mg/dl(true=1, false=0))    
restecg->resting electrocardiographic result->(0->Normal, 1->ST-T Wave abnormality,2->probable or definite left ventricular hypertrophy)    
thalach->max heart beat achieved   
exang->excercise induced angina   
oldpeak->ST depression induced by exercise related to rest   
slope->slope of peak ST segement (1->upsloping, 2->flat, 3->downslopinh)  
ca->number of major vessels   
thal->3->normal, 6->fixed defect, 7->reversible defect   
target->0=less chances of heartattack, 1=more chances of heartattack

# Gender Distribution acc to the target variable

```python
sns.countplot(data=df,x='sex',hue='target')
plt.xticks([0,1],['Female','Male'])
plt.legend(labels=['No','Yes'])
```

# Age distribution in given dataset

```python
sns.distplot(df['age'],bins=30)
```

So, Most people are in 50-60 age group in this dataset 

```python
labels = ["{0} - {1}".format(i, i + 20) for i in range(1, 118, 20)]

df['Age_group'] = pd.cut(df.age, range(1, 130, 20), right=False, labels=labels)
sns.countplot(data=df,x='Age_group',hue='target')
plt.legend(labels=['No','Yes'])
```

People in age group 41-61 have higher chances of having a heart disease

```python
sns.lineplot(data=df,x='age',y='chol')
plt.show()
```

People in 55-70 age group usually have higher cholestrol level

# Chest Pain Distribution according to target variable 

```python
sns.countplot(data=df,x='cp',hue='target')
plt.xticks([0,1,2,3],['typical angina','atypical angina','non-anginal','asymptomatic'])
plt.legend(labels=['No','Yes'])
plt.show()
```

People with non-anginal chest pain are more prone to heart disease
